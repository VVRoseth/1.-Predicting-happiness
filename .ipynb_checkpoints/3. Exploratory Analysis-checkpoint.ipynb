# Predicting happiness 

## Exploring correlations in clean dataset 

### 1. Imports and dataset:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the clean and dummied dataset in
colstep = pd.read_csv('STEPColombia.csv')
```

```python
# Confirming this is the correct dataset 
colstep.head()
```

### 2. Identifying correlations:

The dataset includes variables that can be grouped under seven categories:

1. Health
2. Wealth
3. Personality traits
4. Education
5. Labor market status and work
6. Other personal characteristics
7. Family composition

I will explore the correlation between the variables under each category with each other to identify possible colinearity and with life satisfaction to explore which factors have the strongest correlation with life satisfaction

**2.1. Correlation - Health:**

```python
# Creating a dataframe with health features 
health_features = colstep[['life_satisfaction','offdays_ill', 'healthinsurance', 'chronic_disease', 'BMI']]
```

```python
# Looking at the correlation matrix. There is a strong, positive, correlation between chronic
# disease and days off due to illness, which makes sense. 
matrix = np.triu(health_features.corr())
plt.title("Correlation among health features")
sns.heatmap(health_features.corr(), annot = True, cmap = 'Oranges', mask=matrix);
```

```python
# Getting a closer look at the correlation of health features and life satisfaction. Correlations are not strong. 
plt.figure(figsize=(2,5))
plt.title("Correlation between health and life satisfaction")
sns.heatmap(health_features.corr()[['life_satisfaction']], annot = True, cmap = 'Oranges');
```

**2.2. Correlation - Wealth:**

```python
# Creating a dataframe with wealth features 
wealth_features = colstep[['life_satisfaction','owns_house', 'house_beds', 'house_kitchen', 'reported_social_status', 'got_pu_transf', 
                           'got_pr_transf', 'shocks_bef_15', 'wealth_index']]
```

```python
# Looking at the correlation matrix. The correlation between wealth features is mostly weak, which is positive. The only risk of colinearity comes from 
# belongings, reported social status and the wealth index. This makes sense as the index is partially created from those variables. I may need to choose the
# wealth index over the other features to limit covariance. 

matrix = np.triu(wealth_features.corr())
plt.figure(figsize=(8,6))
plt.title("Correlation among wealth features")
sns.heatmap(wealth_features.corr(), annot = True, cmap = 'Oranges', mask=matrix);
```

```python
# Getting a closer look at the correlation of wealth features and life satisfaction. The correlation between the wealth index and life satisfaction is the 
# largest one we have found so far (0.19)
plt.figure(figsize=(2,7))
plt.title("Correlation between wealth and life satisfaction")
sns.heatmap(wealth_features.corr()[['life_satisfaction']], annot = True, cmap = 'Oranges');
```

**2.3. Correlation - Personality Traits:**

```python
# Creating a dataframe with personality trait features 
ptraits_features = colstep[['life_satisfaction','extraversion_av', 'conscientiousness_avg', 'openness_av', 'stability_av', 'agreeableness_av', 'grit_av', 
                            'decision_av', 'hostile_av', 'risk']]
```

```python
# Looking at the correlation matrix. There are a few personality traits that correlate to each other, which is to be expected. Correlations are
# not strong enough to fear colinearity. 

matrix = np.triu(ptraits_features.corr())
plt.figure(figsize=(8,6))
plt.title("Correlation among personality trait features")
sns.heatmap(ptraits_features.corr(), annot = True, cmap = 'Oranges', mask=matrix);
```

```python
# Getting a closer look at the correlation of personality traits and life satisfaction. The correlation between the stability score and life satisfaction (0.18)
# is the largest in this category. 
plt.figure(figsize=(2,7))
plt.title("Correlation between personality traits and life satisfaction")
sns.heatmap(ptraits_features.corr()[['life_satisfaction']], annot = True, cmap = 'Oranges');
```

**2.4. Correlation - Education/Skills:**

```python
# Creating a dataframe with education features 
education_features = colstep[['life_satisfaction','in_school', 'part_in_training', 'read_overall', 'write_overall', 'numeracy_overall', 'computer_use_overall',
                              'highest_ISCED_PIAAC','years_educ', 'overqualified', 'dropout']]
```

```python
# Looking at the correlation matrix. The heatmap shows that there is colinearity in various features. The most prominent are computer use/overqualification 
# with years of education and educational attainment. The high correlation between educational attainment and years of education is obvious. I will not discard
# any as  these are key variables to guarantee comparability accross countries. 

matrix = np.triu(education_features.corr())
plt.figure(figsize=(8,6))
plt.title("Correlation among education/skills features")
sns.heatmap(education_features.corr(), annot = True, cmap = 'Oranges', mask=matrix);
```

```python
# Getting a closer look at the correlation of education and life satisfaction. The correlations are generally weak. 

plt.figure(figsize=(2,7))
plt.title("Correlation between education/skills and life satisfaction")
sns.heatmap(education_features.corr()[['life_satisfaction']], annot = True, cmap = 'Oranges');
```

**2.5. Correlation - Labor market status / work:**

```python
# Creating a dataframe with those features 

lms_features = colstep[['life_satisfaction','supervise', 'think_learn_work', 'autonomy_at_work', 'repetitiveness_at_work', 'physical_demand_work', 
                        'labor_market_status', 'job_stable', 'hour_earnings', 'total_hr_worked_week']]
```

```python
# Looking at the correlation matrix. This correlation matrix shows that these features have colinearity, which will need to be contolled during the 
# modeling phase

matrix = np.triu(lms_features.corr())
plt.figure(figsize=(8,6))
plt.title("Correlation among labor market statys / work features")
sns.heatmap(lms_features.corr(), annot = True, cmap = 'Oranges', mask=matrix);
```

```python
# Getting a closer look at the correlation of labor market satus / work and life satisfaction. The correlations are generally weak. 

plt.figure(figsize=(2,7))
plt.title("Correlation between labor market status / work and life satisfaction")
sns.heatmap(lms_features.corr()[['life_satisfaction']], annot = True, cmap = 'Oranges');
```
